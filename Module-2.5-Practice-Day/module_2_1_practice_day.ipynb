```python
# Part A – Central Tendency and Spread
# 1. Marks: 50, 60, 65, 70, 75, 80, 85, 90
# Find mean, median, mode.
# Replace 90 with 900 and recompute mean. Explain the effect of outliers.


```

```python
import pandas as pd
df = pd.DataFrame(
    {
        "Marks":[50, 60, 65, 70, 75, 80, 85, 90]
    }
)

mean_before = df['Marks'].mean()
median_before = df['Marks'].median()
mode_before = df['Marks'].mode()
print(mean_before,median_before,mode_before)
```

```python
df_outlier = pd.DataFrame({
    "Marks": [50, 60, 65, 70, 75, 80, 85, 900]
})
mean_after = df_outlier['Marks'].mean()
median_after = df_outlier['Marks'].median()
mode_after = df_outlier['Marks'].mode()
print(mean_after,median_after,mode_after)

#Outlier (যেমন 900) mean-কে অনেক বাড়িয়ে দেয় কারণ mean সব মানের যোগফলের উপর হিসাব হয়।

# আগে Mean ছিল কম (71.875 এর কাছাকাছি), 
# Outlier যোগ হলে Mean লাফ দিয়ে অনেক বেশি হয়ে যায় (170+).

# Median একই রকম থাকে কারণ median শুধু মাঝের মান দেখে।

```

```python
# 2. Temperatures (°C): 29, 31, 33, 33, 32, 31, 30
# Calculate variance and standard deviation.
# Which measure is easier to interpret and why?


df = pd.DataFrame({
    "Temp": [29, 31, 33, 33, 32, 31, 30]
})

mean = df["Temp"].mean()
variance = df["Temp"].var(ddof=0)   
std_dev = df["Temp"].std(ddof=0)   

mean, variance, std_dev

print("Mean =", mean)
print("Variance =", variance)
print("Standard Deviation =", std_dev)
```

```python
import pandas as pd


weights = [2.5, 3.0, 3.2, 3.3, 3.4, 3.5, 3.6, 3.9, 4.0, 4.5]

df = pd.DataFrame({"Weight":weights})

# Percentiles
P25 = df["Weight"].quantile(0.25)
P50 = df["Weight"].quantile(0.50)  # Median
P75 = df["Weight"].quantile(0.75)

IQR = P75 - P25

lower_bound = P25 - 1.5 * IQR 
upper_bound = P25 + 1.5 * IQR

outliers = df[(df["Weight"] < lower_bound) | (df["Weight"] > upper_bound)]

print("\nPossible Outliers:")
print(outliers if not outliers.empty else "No outliers detected")

```

```python
# Module 2 – Probability Basics for ML
# Part A – Basic Probability
# 1. In a survey of 100 people:
# 40 like pizza (A)
# 50 like burgers (B)
# 20 like both (A ∩ B)
#  Find P(A), P(B), P(A ∩ B), and P(A | B).

import pandas as pd

total_people = 100
like_pizza = 40
like_burger = 50
like_both = 20

P_A = like_pizza / total_people
P_B = like_burger / total_people
P_A_and_B = like_both / total_people
P_A_given_B = P_A_and_B / P_B

print(f"P(A) = {P_A}")
print(f"P(B) = {P_B}")
print(f"P(A ∩ B) = {P_A_and_B}")
print(f"P(A | B) = {P_A_given_B}")

```

```python
# Part B – Conditional Probability and Bayes
# 2. Email dataset of 1 000 mails:
# 100 spam (S)
# 40 of these contain “free” (F)
# 60 non-spam also contain “free”
#  Compute P(S | F) using Bayes’ Theorem →
 
#  Explain what this means for a spam filter.


# 3. Disease example:
# P(Disease) = 1 / 10 000
# P(+ | D) = 0.99
# P(− | ¬D) = 0.99
 
# Find P(D | +) and discuss the base-rate effect.

# Given data
total_mails = 1000
spam = 100
free_in_spam = 40
non_spam = total_mails - spam
free_in_non_spam = 60

# Calculate probabilities
P_S = spam / total_mails
P_not_S = non_spam / total_mails
P_F_given_S = free_in_spam / spam
P_F_given_not_S = free_in_non_spam / non_spam

# Total probability of free (F)
P_F = P_F_given_S * P_S + P_F_given_not_S * P_not_S

# Bayes Theorem: P(S|F)
P_S_given_F = (P_F_given_S * P_S) / P_F

print(f"P(S|F) = {P_S_given_F:.4f}")
```

```python
# Given data
P_D = 1 / 10000
P_not_D = 1 - P_D
P_pos_given_D = 0.99
P_neg_given_not_D = 0.99
P_pos_given_not_D = 1 - P_neg_given_not_D  # False positive

# Bayes Theorem: P(D|+)
P_D_given_pos = (P_pos_given_D * P_D) / ((P_pos_given_D * P_D) + (P_pos_given_not_D * P_not_D))

print(f"P(Disease|Positive) = {P_D_given_pos:.4f}")
```
